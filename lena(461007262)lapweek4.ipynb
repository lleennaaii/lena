```python
total = 0

for i in range(1, 11):
    num = int(input(f"Enter number {i}: "))

    if num == 0:
        print("Input stopped by user.")
        break # Correctly indented break statement

    if num < 0:
        continue

    if num % 2 == 0:
        total += num # augmented assignment

print("Sum of positive even numbers:", total)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 13):
    product = num * i
    if product > 50:
        break
    if product % 3 == 0:
        continue
    print(f"{num} x {i} = {product}")
```

```python
secret_number = 7
attempts = 0

while True:
    guess = int(input("Guess the number (1-20): "))

    if guess < 1 or guess > 20:
        print("Invalid guess, try again.")
        continue  # skip invalid guess

    attempts += 1  # count valid attempts

    if guess == secret_number:
        print(f"Congratulations! You guessed it in {attempts} attempts.")
        break
    elif guess < secret_number:
        print("Too low.")
    else:
        print("Too high.")
```

```python
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):
    if i > 50:
        break
    if i % 8 == 0:
        continue
    if i % 4 == 0:
        count += 1  # augmented assignment

print("Count of numbers divisible by 4 (excluding multiples of 8, stopping at >50):", count)
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        if j == 3:
            continue
        if i * j > 6:
            break
        print(j, end=" ")
    print()
```
