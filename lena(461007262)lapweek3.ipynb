```python
count=0
total=0
num=int(input("Enter a number(-999 to quit) :"))
while num != -999:
    count +=1
    total+=num
    num=int(input("Enter a number(-999 to quit) :"))
if count==0:
   print("No Numbers were entered")
else:
   average=total/count
   print("Number of entrise :",count)
   print("Sum :",total)
   print("Average:",average)
```
