```python
import math

def circle_area(radius):
    return math.pi * radius ** 2

circle_circumference = lambda r: 2 * math.pi * r

radius = float(input("Enter the radius of the circle: "))

area = circle_area(radius)
circumference = circle_circumference(radius)

print(f"Area: {area:.2f}")
print(f"Circumference: {circumference:.2f}")
```

```python
import re

def sum_numbers(numbers):
    return sum(numbers)

double = lambda x: x * 2

text = input("Enter a string: ")

numbers_found = list(map(int, re.findall(r'\d+', text)))

total = sum_numbers(numbers_found)
doubled = double(total)

print("Numbers found:", numbers_found)
print("Sum of numbers:", total)
print("Doubled sum:", doubled)
```

```python
import math

def squares(numbers):
    sq = list(map(lambda x: x**2, numbers))
    return sq

nums = input("Enter numbers separated by commas: ")
num_list = list(map(float, nums.split(',')))

squared_list = squares(num_list)
max_square = max(squared_list)
sqrt_sum_squares = math.sqrt(sum(squared_list))

print("Squares:", squared_list)
print("Maximum square:", max_square)
print("Square root of sum of squares:", sqrt_sum_squares)
```

```python
import re
import math

def password_strength(pw):
    if (re.search(r'[A-Z]', pw) and
        re.search(r'[a-z]', pw) and
        re.search(r'\d', pw)):
        return "Strong"
    else:
        return "Weak"

average = lambda scores: sum(scores)/len(scores)

password = input("Enter a password: ")
scores = list(map(float, input("Enter 5 scores separated by space: ").split()))

strength = password_strength(password)
avg_score = average(scores)
rounded_avg = math.ceil(avg_score)

print("Password strength:", strength)
print("Rounded average score:", rounded_avg)
```

```python
import re
import math

def count_vowels(text):
    return len(re.findall(r'[aeiouAEIOU]', text))

word_lengths = lambda words: list(map(len, words))

paragraph = input("Enter a paragraph of text: ")

vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

total_vowels = count_vowels(paragraph)

lengths = word_lengths(vowel_words)
avg_length = round(sum(lengths)/len(lengths), 2) if lengths else 0

print("Words starting with vowels:", vowel_words)
print("Total vowels:", total_vowels)
print("Average length of vowel-starting words:", avg_length)
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

try:
    num = int(input("Enter a non-negative integer: "))
    if num < 0:
        raise ValueError("Negative number not allowed")
    print(f"Factorial of {num} is {factorial(num)}")
except ValueError as e:
    print("Invalid input:", e)
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

try:
    n_terms = int(input("Enter the number of Fibonacci terms: "))
    if n_terms <= 0:
        raise ValueError("Number must be positive")
    print("Fibonacci sequence:")
    for i in range(n_terms):
        print(fibonacci(i), end=" ")
except ValueError as e:
    print("Invalid input:", e)
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    elif exp > 0:
        return base * power(base, exp-1)
    else:
        return 1 / power(base, -exp)

try:
    base = float(input("Enter the base: "))
    exponent = int(input("Enter the exponent: "))
    print(f"{base} raised to {exponent} is {power(base, exponent)}")
except ValueError as e:
    print("Invalid input:", e)
except ZeroDivisionError as e:
    print("Error:", e)
```
