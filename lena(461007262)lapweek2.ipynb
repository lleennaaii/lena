```python
#ask user to enter name
name = input("Enter your name:")
#ask user to enter age
age = input("Enter your age:")
#print greeting message
print("Hello", name, "you are", age, "years old.")
```

```python
#take number from user
num = float(input("Enter a number:"))
#check if number is positive
if num > 0:
  print("The number is positive.")
  #check if number is negative
elif num < 0:
  print("The number is negative.")
  #if number is zero
else:
  print("The number is zero")
```

```python
#take marks from user
marks = int(input("Enter your marks (0-100):"))
#check grads condition
if marks >= 85:
  print("Grade: A")
elif marks >= 70:
  print("Grade: B")
elif marks >= 50:
  print("Grade: C")
  #if marks less than 50
else:
  print("Fail")
```

```python
#ask how many numbers user will enter
count = int(input("How many numbers do you want to enter?"))
#initialize total to zero
total = 0
#loop to take numbers
for i in range(count):
  num = float(input("Enter a number:"))
  total = total + num
#calculate average
  average = total / count
print("Sum =", total)
print("Avarage =1", average)
```

```python
#function to calc area of circle
def calculate_area(radius):
    return  3.14 * radius * radius
 #take radius from user
r = float(input("Enter the radius of the circle:"))
#call function
area = calculate_area(r)
print("The area of the circle is:", area)
```

```python
#keep programing running
while True:
  #display menu
  print("\n1. Add two numbers")
  print("2. Subtract two numbers")
  print("3. Exit")
  #take user choice
  choice = input("Enter your choice:")
  #addition
  if choice == "1":
    a=float(input("Enter first number:"))
    b= float(input("Enter second number:"))
    print("Result =", a+b)
    #subtraction
  elif choice =="2":
    a= float(input("Enter first number:"))
    b= float(input("Enter second number:"))
    print("Result =", a-b)
    #exit program
  elif choice == "3":
    print("Goodbye!")
    break
  else:
    print("invalid choice")
```

```python
#initialize total expense
total = 0
#loop to take expense
while True:
  expense = float(input("Enter expense (0 to stop):"))
  #stop if user enters 0
  if expense == 0:
    break
    #add expense to total
    total = total + expense
print("Total expenses =", total)
```

```python
#predefined username and password
correct_username ="admin"
correct_password = "1234"
#counter for attempts
attempts = 0
#allow only 3 attempts
while attempts < 3:
  #take user input
  username = input("Enter username:")
  password = input("Enter password:")
  #check credentials
  if username == correct_username and password == correct_password:
    print("Login successful!")
    break
  else:
    print("Invalid username or password.")
    attempts = attempts + 1
    #if all attemots used
    if attempts == 3:
      print("Account Locked")
```
