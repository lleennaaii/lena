```python
# List to store songs
playlist = []

# Function to add a song
def add_song():
    name = input("Enter song name: ")
    artist = input("Enter artist name: ")
    album = input("Enter album name: ")

    song = {
        "name": name,
        "artist": artist,
        "album": album
    }

    playlist.append(song)
    print("Song added successfully! 🎶\n")


# Function to display playlist
def display_playlist():
    if not playlist:
        print("Playlist is empty.\n")
        return

    print("\n🎵 Your Playlist:")
    for i, song in enumerate(playlist, start=1):
        print(f"{i}. {song['name']} - {song['artist']} ({song['album']})")
    print()


# Function to search song
def search_song():
    search_name = input("Enter song name to search: ")

    found = False
    for song in playlist:
        if song["name"].lower() == search_name.lower():
            print("🎧 Song Found:")
            print(f"{song['name']} - {song['artist']} ({song['album']})\n")
            found = True
            break

    if not found:
        print("Song not found.\n")


# Main menu
def main():
    while True:
        print("===== 🎵 Music Playlist Manager =====")
        print("1. Add Song")
        print("2. Display Playlist")
        print("3. Search Song")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_song()
        elif choice == "2":
            display_playlist()
        elif choice == "3":
            search_song()
        elif choice == "4":
            print("Goodbye 👋")
            break
        else:
            print("Invalid choice. Try again.\n")


# Run program
main()
```
