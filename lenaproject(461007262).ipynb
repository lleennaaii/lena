```python
# Movie Manager
# This program allows user to add, display, and search movie records

# List to store movie records
movies = []

# Function to add a new movie
def add_record():
    name = input("Enter movie name: ")
    genre = input("Enter genre: ")
    rating = float(input("Enter rating (1-10): "))

    movie = {
        "name": name,
        "genre": genre,
        "rating": rating
    }

    movies.append(movie)
    print("Movie added successfully!\n")


# Function to display all movies
def display_records():
    if not movies:
        print("No movies found.\n")
        return

    print("\nMovie List:")
    for i, movie in enumerate(movies, start=1):
        print(f"{i}. Name: {movie['name']}, Genre: {movie['genre']}, Rating: {movie['rating']}")
    print()


# Function to search for a movie
def search_record():
    search_name = input("Enter movie name to search: ")

    found = False
    for movie in movies:
        if movie["name"].lower() == search_name.lower():
            print("Movie Found:")
            print(f"Name: {movie['name']}, Genre: {movie['genre']}, Rating: {movie['rating']}\n")
            found = True
            break

    if not found:
        print("Movie not found.\n")


# Main menu
def main():
    while True:
        print("===== Movie Manager =====")
        print("1. Add Movie")
        print("2. Display Movies")
        print("3. Search Movie")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_record()
        elif choice == "2":
            display_records()
        elif choice == "3":
            search_record()
        elif choice == "4":
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Try again.\n")


# Run program
main()
```
